```python
#Create a new set of 2000 random latitudes and longitudes

#import dependency for random numbers and numpy
import random
import numpy as np
import pandas as pd

#get 2000 random numbers using uniform and assign to lat or lng list
lats=np.random.uniform(-90.000, 90.000, size=2000)
lngs=np.random.uniform(-90.000, 90.000, size=2000)
```

```python
#import the citipy module
from citipy import citipy

#combine the individual lats and lngs
lat_lngs=zip(lats, lngs)

#put into a list
coordinates=list(lat_lngs)

#initialize a cities list to put the nearest cities into
cities=[]

#identify nearest city for each lat lng combo and append to the list if not already in the list
#for each lat-lng combo:
for coordinate in coordinates:
    #use citipy to find a city based on the lat-lng combo and get the city's name
    city=citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    
    #if city is not already in list, add to list
    if city not in cities:
        cities.append(city)

#check the length of cities list
len(cities)
```

```python
#perform an api call with openweathermap and retreive data

#import the dependencies for requests and the api key
import requests
from config import api_key2

#build the base url for openweathermap
basic_url="http://api.openweathermap.org/data/2.5/weather?units=imperial&appid="+api_key2
```

```python
#practice getting each data
#prac_url=requests.get(basic_url+"&q="+cities[1]+"&lang=en").json()
#prac_url["weather"][0]["description"]
```

```python
#initialize list for city data
cities_data=[]

for city in cities:
    #get url for the city
    url=basic_url+"&q="+ city.replace(" ", "+")
    
    try:
        #get the url and convert to json()
        city_url=requests.get(url).json()
        #print(city_url["coord"])
        
        #get data
        lat=city_url["coord"]["lat"]
        lng=city_url["coord"]["lon"]
        max_temp=city_url["main"]["temp_max"]
        humidity=city_url["main"]["humidity"]
        cloudiness=city_url["clouds"]["all"]
        wind_speed=city_url["wind"]["speed"]
        desc=city_url["weather"][0]["description"]
        country=city_url["sys"]["country"]
        
        #append to cities_data list
        cities_data.append({
            "City":city,
            "Country":country,
            "Lat": lat,
            "Lng": lng,
            "Max Temp": max_temp,
            "Humidity": humidity,
            "Cloudiness":cloudiness,
            "Wind Speed": wind_speed,
            "Current Description": desc
        })
    #skip cities that can't be found
    except:
        print("skip")
        pass
```

```python
#add the data to a new dataframe
city_df=pd.DataFrame(cities_data)
city_df
```

```python
#export the dataframe as a csv file
#create output file
output_data_file="WeatherPy_Database.csv"

#export
city_df.to_csv(output_data_file)
```
